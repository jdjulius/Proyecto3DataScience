# Ejercicio Nº 4
Crear una función que muestre los siguientes gráficos (simultáneamente) para las variables numéricas, según una variable cualitativa:

1. Cajas y bigotes
2. Barras de la media
3. Histograma con densidad (Curva)

La función debe contar con su respectiva documentación.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/autos.csv", delimiter=";")
df
```

```python
def mapa_calor_por_pais(df, varCualitativa):

    variables = df.select_dtypes(include=['int64', 'float64'])

    agrupacion = df.groupby(varCualitativa)[variables.columns].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(agrupacion, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

    plt.title('Correlación de Variables Numéricas por País')
    plt.xlabel('Variables Numéricas')
    plt.ylabel('Variables por países')

    return plt.gcf()
```

```python
mapa_calor = mapa_calor_por_pais(df, 'Pais_Origen')
plt.show(mapa_calor)
```
